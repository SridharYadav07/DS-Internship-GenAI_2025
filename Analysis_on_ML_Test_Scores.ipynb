**Analyzing the Data**

```python
data.columns = data.columns.str.strip()
batch_stats = data.groupby('Batch')['Score'].describe()
batch_stats
```

**Score Distribution for Each Batch**

```python
sns.set(style="whitegrid")
```

```python
##### Creating a boxplot to compare score distributions by batch

plt.figure(figsize=(10, 6))
sns.boxplot(x='Batch', y='Score', data=data)
plt.title("Score Distribution by Batch")
plt.show()
```

##### From the above boxplot, here are some insights regarding the score distribution for each batch:

AI_ELITE_4: This batch shows a relatively lower performance, with the median score appearing to be around 4 or 5 out of 7. The socres are concentrated in the mid-range.
AI_ELITE_6: This batch has a slightly better median score than AI_ELITE_4, likely around 5 out of 7. The scores are also more spread out, indicating a wider range of performance within this batch.
AI_ELITE_7: This batch exhibits the highest performance, with the median score being around 6 or 7 out 7. The scores are also more tightly clustered at the higher end, suggesting more consistent high performance from this group.

**Create a histogram to visualize the distribution of scores**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Score', hue='Batch', multiple="stack", kde=True)
plt.title("Score Distribution across Batches")
plt.show()
```

##### From the above histogram titled 'Score Distribution across Batches', here are the key insights:

Overall Score Range: The scores generally range from 1/7 to 7/7 across all batches.
AI_ELITE_4: This batch shows a distibution with a higher frequency of lower socres, particularly around 3/7 and 4/7, and a gradual decrease towards higher socres.
AI_ELITE_6: This batch's scores appear to be more spread out, with a peak around 5/7 and also notable frequencies at 4/7 and 6/7. It shows a wider range of performance compared to AI_ELITE_4.
AI_ELITE_7: This batch clearly has the highest performance, with a strong concentration of scores at 6/7 and 7/7. The distribution is skewed towards the higher end, indicating consistent high achievement.
