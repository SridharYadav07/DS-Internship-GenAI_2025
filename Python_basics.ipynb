**Python If-Else**

```python
n = int(input().strip())
if n % 2 == 1:
  print("Weird")
elif 2 <= n <=5:
  print("Not Weird")
elif 6 <= n <=20:
  print("Weird")
else:
  print("Not Weird")
```

**Python Division**

```python
a = int(input())
b = int(input())

print(a // b)
print(a / b)
```

**Python Loops**

```python
n = int(input())

for i in range(n):
  print(i * i)
```

**Python Function**

```python
def is_leap(year):
    leap = False

    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        leap = True

    return leap

year = int(input())
print(is_leap(year))
```
