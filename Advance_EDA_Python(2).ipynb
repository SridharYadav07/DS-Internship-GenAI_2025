**Defanging an IP Address**
####Given a valid (IPv4) IP address, return a defanged version of that IP address. A defanged IP address replaces every period "." with "[.]"

```python
def defangIPaddress(address: str) -> str:

  return address.replace('.', '[.]')

print(defangIPaddress("1.1.1.1"))
```

 **Find Numbers with Even Number of Digits** **bold text**

#####Given an array nums of integers, return how many of them contain an even number of digits

```python
def countNumberWithEvenDigits(nums):
  count = 0

  for num in nums:
    if len(str(num)) % 2 == 0:
      count += 1

  return count

print(countNumberWithEvenDigits([12, 345, 2, 6, 7896]))
```

**Number of Good Pairs**
######Given an array of integers nums, return the number of good pairs. A pair(i, j) is called good if nums[i] == nums[j] and i < j.

```python
def numIdenticalPairs(nums):
  good_pairs = 0

  for i in range(len(nums)):
    for j in range(i + 1, len(nums)):

      if nums[i] == nums[j]:
        good_pairs +=1

  return good_pairs

print(numIdenticalPairs([1, 2, 3, 1, 1, 3]))
```

**How Many Numbers Are Smaller Than the Current Number**
######Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. That is, for each nums[i] you have to count the number of valid j's such that j !=i and nums[j] < nums[i]. Return the answer in an array.

```python
def smallerNumbersThanCurrent(nums):
  result = []

  for i in range(len(nums)):
    count = 0

    for j in range(len(nums)):
      if nums[j] < nums[i]:
        count += 1

    result.append(count)

  return result

print(smallerNumbersThanCurrent([8, 1, 2, 2, 3]))
```

**Subtract the Product and Sum of Digits of an Integer**
######Given an integer number n, return the difference between the product of its digits and the sum of its digits.

```python
def subtractProductAndSum(n):

  product = 1
  sum_digits = 0

  for digit in str(n):
    digit = int(digit)
    product *= digit
    sum_digits += digit

  return product - sum_digits

print(subtractProductAndSum(234))
```

**XOR Operation in an Array**
###### You are given an integer n and an integer start. Define an array nums where nums[i] = start + 2 * i (0-indexed) and n == nums.length. Return the bitwise XOR of all elements of nums.


```python
from re import IGNORECASE
def xorOperation(n, start):

  result = 0

  for i in range(n):
    value = start + 2 * i
    result ^= value

  return result

print(xorOperation(5, 0))
```
